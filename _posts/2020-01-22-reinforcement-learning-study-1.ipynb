# 6. 심층 순방향 신경망
- 심층 순방향 신경망 = 순방향 신경망 or 다층 퍼셉트론
- 목표: 어떤 함수 $f^*$를 근사하는 것.
- 하나의 사상 $\vy=f(\vx;\vtheta)$를 정의하고, 실제 목표 함수를 가장 잘 근사하는 매개변수 $\vtheta$의 값들을 학습한다.

뇌에 비유하기보다는, 통계적 일반화를 위해 고안된 함수 근사 기계로 간주하는것이 더 낫다.

선형 모형으로는 비선형 모형을 나타낼 수 없기 때문에 $\vx$에 대한 비선형 함수 모형을 만들기 위해 $\phi$라는 비선형 함수를 도입하여, 선형 모형을 $\phi(\vx)$에 적용시킨다. 선형 모형이 학습할 수 없는 대표적인 예는 XOR 함수가 있다.

심층학습에서는 이 $\phi$를 학습하게 한다. 이때의 모형은 $y=f(\vx;\vtheta,\vw)=\tr{\phi(\vx;\vtheta)}\vx$ 이다.

## 6.1 XOR의 학습

선형 모형이 적절하지 않은 예. 원본 함수 $y=f(\vx;\vtheta)$를 최대한 근사하도록 $\vtheta$를 학습하여 근사 함수 $f^*$를 구하는 것이 목적.
학습 방법: 평균제곱오차(MSE) (보통은 신경학습에 적절하지 않다고 한다.)
전체 훈련 집합에 대한 MSE를 이용한 손실 함수:
$$J(\vtheta)=\frac{1}{4}\sum_{\vx \in \sX}(f^*(\vx)-f(\vx;\vtheta))^2 \tag{6.1}$$
근데 선형 모형으로 두게 되면, 모형을 다음과 같이 둘 수 있다.
$$f(\vx;\vw,b)=\tr{\vx}\vw + b \tag{6.2}$$
이 표준방정식 (normal equation; 또는 정규방정식, 5장 식 5.12 참조, 그러나 5.12의 식은 원점을 지나는 경우에 대한 것임.)을 풀면 $\vw=0$과 $b=\frac{1}{2}$를 얻는다. (?? 어떻게 푼거지??)

이런 문제를 해결하기 위해 심층학습을 사용하는데, 여러 층을 쓰더라도 전부 선형 계산이면 결국 선형 모형이 되어 의미가 없다. 그래서 활성화 함수라 불리는 고정된 비선형 함수를 사용한다. 그중 정류 선형 단위(rectified linear unit, ReLU)를 많이 쓴다. $g(z)=maz\{0,z\}$
(??왜 ReLU가 좋은걸까. 이제껏 계산한걸 마이너스이기만 하면 다 버려버리는데. 그러면 매개변수 설정에 따라 반정도는 버리고 가는거 아닌가 싶은데.. 아직 잘 이해를 못한 듯 하다)
아래는 ReLU함수의 그래프. 선형에 가까워서 선형 모형의 성질을 많이 가지고 있다고 한다.



```python
import unittest
import numpy as np
import numpy.testing as nt
from matplotlib import pyplot as plt
%matplotlib inline
x=np.arange(-2,2, 0.1)
y=x* (x>0)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('ReLU')
```

$$f(\vx;\mW,\vc,\vw,b)=\tr{\vw} max\{{0,\tr{\mW} \vx+c}\}+b \tag{6.3}$$

```python
%matplotlib inline

import numpy as np

pdf=np.random.normal
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

abs(mu - np.mean(s)) < 0.01

abs(sigma - np.std(s, ddof=1)) < 0.01

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 4차원 함수를 써보자.

def f(x):
    return 0.4*x**4-8*x**3+20*x**2-42*x**1+150

def gradi(x):
    return 4*0.4*x**3-24*x**2+40*x**1-42


range1 = np.arange(-10,20,0.1)
ys=f(range1)
funArray=np.array([range1,ys])
plt.plot(range1, ys)
```

```python
# 초기 x를 설정
x=-10
# gradient
costs=[]
# m denotes the number of examples here, not the number of features
def gradientDescent(x, y, alpha, m, numIterations):
    global costs
    xTrans = x.transpose()
    cost=999
    grad=np.gradient(funArray, axis=-1)
    for i in range(0, numIterations):
        hypothesis=x
        loss = hypothesis - f(x)
        loss=f(x)-f(13)
        # avg cost per example (the 2 in 2*m doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        cost = np.sum(loss ** 2) / (2 * m)
        costs.append(cost)
        #gradient = np.dot(xTrans, loss) / m
        gradient=gradi(x)
        
        # update
        x=x-alpha*gradient
        if(i%100==0):
            print("Iteration %d | Cost: %f" % (i, cost))
            # avg gradient per example
            print(hypothesis)
    print(cost)
    return x

learning_rate=0.001
epoch=300
x=np.array([-10])
m=len(range1)
x=gradientDescent(x,ys,learning_rate,m,epoch)
print(x)
plt.plot(range1,costs)
plt.show()
costs=np.array(costs)


fig, ax1 = plt.subplots()

grad=np.gradient(funArray, axis=-1)
#plt.plot(range1,grad[1])

color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('grad', color=color)
ax1.plot(range1, grad[1], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('cost', color=color)  # we already handled the x-label with ax1
ax2.plot(range1, costs, color=color)
ax2.tick_params(axis='y', labelcolor=color)

```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot
import matplotlib.pyplot as plt
import random

seed=0
def fun(x, y):
    global seed
    seed+=1
    return x**2 + y**2 -2*y +3

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.5)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
grad=np.gradient(zs)

Z = zs.reshape(X.shape)
GRAD=grad.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,GRAD)
```
